# Vector-Ángulo y Cuaterniones

Este cuaderno explora representaciones de rotación en 3D usando:

- **Vector-ángulo** (eje + ángulo) con la fórmula de Rodrigues.
- **Cuaterniones**, y cómo convertir entre cuaterniones y matrices de rotación.

La idea es ver cómo se construyen/recuperan rotaciones y cuáles son las ventajas numéricas de cada forma.

```python
import numpy as np
```

```python
def skew(k):
    # Matriz antisimétrica asociada al producto cruz (k x v)
    kx, ky, kz = k
    return np.array([
        [0,  -kz,  ky],
        [kz,  0,  -kx],
        [-ky, kx,  0]
    ])

def dir_axang(k, theta):

    k = np.asarray(k).reshape(-1)

    if k.size != 3:
        raise ValueError("k debe ser un vector de R3")

    # Normalizar el eje de rotación
    k = k / np.linalg.norm(k)

    I = np.eye(3)

    # Fórmula de Rodrigues para rotación alrededor de un eje unitario k
    R = (np.cos(theta) * I +
         (1 - np.cos(theta)) * np.outer(k, k) +
         np.sin(theta) * skew(k))

    return R
```

```python
# Versión más estable que usar descomposición en autovectores
def inv_axang(R):    
    # theta = acos((tr(R)-1)/2) viene de la relación entre la traza y el ángulo de rotación
    theta = np.arccos((np.trace(R) - 1) / 2)

    if np.isclose(theta, 0):
        # rotación nula: el eje es arbitrario
        k = np.array([1,0,0])
    else:
        # parte anti-simétrica de R da el eje (normalizado por sin(theta))
        # R-R' = 2 sin(theta) * k
        k = np.array([
            R[2,1] - R[1,2],
            R[0,2] - R[2,0],
            R[1,0] - R[0,1]
        ]) / (2*np.sin(theta))

    return k, theta
```

### Prueba numérica (eje+ángulo)

Se construye una matriz de rotación a partir de un eje y un ángulo, se recupera la representación vector-ángulo, y se compara con la original. También se muestra la relación con los autovalores/autovectores de la matriz de rotación.

```python
k = [0, 0, 1]
theta = np.pi/4

R = dir_axang(k, theta)
print(R)

k_rec, theta_rec = inv_axang(R)
print("Diferencia en eje recuperado: ", k_rec - k)
print("Diferencia en ángulo recuperado: ", theta_rec - theta)

# La descomposición en autovalores/autovectores también da el eje de rotación.
# Para una rotación, los autovalores son (1, exp(iθ), exp(-iθ)).
V, l = np.linalg.eig(R)
print(V)
print(l)
```

## Cuaterniones

Los cuaterniones representan rotaciones con cuatro parámetros y evitan algunas singularidades de los ángulos de Euler (gimbal lock).

```python
def dir_quat(q):
    # Convierte un cuaternión Q = [g, qx, qy, qz] a matriz de rotación
    q = np.asarray(q, dtype=float)
    q = q / np.linalg.norm(q)

    g, qx, qy, qz = q

    # Ángulo de rotación: θ = 2 acos(g)
    theta = 2*np.arccos(g)

    # Si g ≈ ±1, la rotación es cercana a la identidad (eje indefinido)
    s = np.sqrt(1 - g*g)

    if np.isclose(s, 0):
        # rotación nula: eje arbitrario
        k = np.array([1.0, 0.0, 0.0])
    else:
        k = np.array([qx, qy, qz]) / s


    R = dir_axang(k, theta)

    return R
```

Para resolver el problema inverso recordar

$$R(\vec{Q})=\begin{bmatrix} 
2q_x q_x +2g^2-1 & 2q_x q_y -2g q_z & 2q_x q_z +2g q_y \\
2q_x q_y +2g q_z & 2q_y q_y +2g^2-1 & 2q_y q_z -2g q_x \\ 
2q_x q_z -2g q_y & 2q_z q_y +2g q_x & 2q_z q_z +2g^2 -1 
\end{bmatrix}
$$

Luego los componentes del cuaternión se encuentran operando sobre elementos de esta matriz

```python
def inv_quat(R):

    # recuperación del cuaternión a partir de la matriz de rotación
    R = np.asarray(R)
    tr = np.trace(R)

    if tr > 0:
        # Caso numéricamente estable cuando tr>0
        g = 0.5 * np.sqrt(tr + 1)

        qx = (R[2,1] - R[1,2]) / (4*g)
        qy = (R[0,2] - R[2,0]) / (4*g)
        qz = (R[1,0] - R[0,1]) / (4*g)

    else:
        # elegir la diagonal dominante
        i = np.argmax([R[0,0], R[1,1], R[2,2]])

        if i == 0:
            qx = 0.5 * np.sqrt(1 + R[0,0] - R[1,1] - R[2,2])
            g  = (R[2,1] - R[1,2]) / (4*qx)
            qy = (R[0,1] + R[1,0]) / (4*qx)
            qz = (R[0,2] + R[2,0]) / (4*qx)

        elif i == 1:
            qy = 0.5 * np.sqrt(1 - R[0,0] + R[1,1] - R[2,2])
            g  = (R[0,2] - R[2,0]) / (4*qy)
            qx = (R[0,1] + R[1,0]) / (4*qy)
            qz = (R[1,2] + R[2,1]) / (4*qy)

        else:
            qz = 0.5 * np.sqrt(1 - R[0,0] - R[1,1] + R[2,2])
            g  = (R[1,0] - R[0,1]) / (4*qz)
            qx = (R[0,2] + R[2,0]) / (4*qz)
            qy = (R[1,2] + R[2,1]) / (4*qz)

    q = np.array([g, qx, qy, qz])

    return q / np.linalg.norm(q)
```

## Ejercicio

- Encontrar la matriz de rotación que expresa la orientación de la terna 2 respecto de la 0
- Encontrar el cuaternión Q que expresa la misma rotación




```python
# Puedo expresar la orientación como composición de rotaciones
R = dir_axang([0,0,1],np.pi/2) @ dir_axang([1,0,0],np.pi/2)
print("Matriz de rotación R=\n",np.round(R, 6))

print(np.linalg.det(R))
```

```python
# Obtengo el cuaternión

Q = inv_quat(R)
print(Q)
```

```python
# Obtengo el eje de rotación

k,theta = inv_axang(R)
print("Eje de rotación k=",k)
print("Ángulo de rotación theta=",theta)
```

### Visualización de la rotación

La animación muestra cómo va girando la terna de referencia alrededor del eje de rotación `k` según el ángulo `theta`.

```python
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
from IPython.display import HTML

# dibujar una terna
def draw_frame(ax, R):
    origin = np.zeros(3)
    colors = ['r','g','b']
    for i in range(3):
        v = R[:,i]
        ax.quiver(*origin, *v, color=colors[i], linewidth=2)

# dibujar el eje
def draw_axis(ax, k, L=1.5):
    k = k / np.linalg.norm(k)
    p1 = -L * k
    p2 =  L * k
    ax.plot(
        [p1[0], p2[0]],
        [p1[1], p2[1]],
        [p1[2], p2[2]],
        color='k',
        linewidth=1
    )

steps = 50

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

#ax.view_init(elev=40, azim=30)
ax.set_axis_off()

def update(frame):

    ax.cla()

    ax.set_xlim([-1,1])
    ax.set_ylim([-1,1])
    ax.set_zlim([-1,1])

    ax.set_axis_off()

    # frame base
    draw_frame(ax, np.eye(3))

    # dibujar eje
    draw_axis(ax, k)

    # rotación progresiva
    angle = theta * frame/steps

    R = dir_axang(k, angle)

    draw_frame(ax, R)

ani = FuncAnimation(fig, update, frames=steps, interval=50)
plt.close(fig)
HTML(ani.to_jshtml())
```
